```python
# Import the relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Load the data
data = pd.read_csv('supply_chain_data.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
# Description of data
data.describe()
```

```python
# Plotting a pie chart between product types and no of products sold
sns.set(style='whitegrid')
sales_by_type = data.groupby('Product type')['Number of products sold'].sum()
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(sales_by_type, labels=sales_by_type.index, autopct='%1.1f%%', startangle=90)
ax.set_title('Total Number of Products Sold by Product Type')
plt.show()
```

Total number of skincare products sold are closely double the number of products sold in othe categories

```python
# Plotting the graph between product type and revenue generated by them
sns.set(style='whitegrid')
revenue_by_type = data.groupby('Product type')['Revenue generated'].sum()
sns.barplot(x=revenue_by_type.index, y=revenue_by_type.values)
plt.title('Revenue Generated by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Revenue')
plt.xticks(rotation=90)
plt.show()
```

Skincare products give the highest revenue and haircare and cosmetics give almost the same but cosmetics being the lowest in revenue

```python
# Plotting a graph between SKU and Revenue Generated
sns.set(style='whitegrid')
fig, ax = plt.subplots(figsize=(24, 10))
sns.lineplot(x='SKU', y='Revenue generated', data=data, ax=ax)
plt.xticks(rotation=90)
ax.set_title('Revenue Generated by SKU')
ax.set_xlabel('SKU')
ax.set_ylabel('Revenue Generated')
plt.show()





```

```python
# Plotting a graph between SKU and Stock Levels
sns.set(style='whitegrid')
fig, ax = plt.subplots(figsize=(24, 10))
sns.lineplot(x='SKU', y='Stock levels', data=data, ax=ax)
plt.xticks(rotation=90)
ax.set_title('Stock levels by SKU')
ax.set_xlabel('SKU')
ax.set_ylabel('Stock Levels')
plt.show()


```

```python
# Plotting a graph between SKU and lead times
sns.set(style='whitegrid')
fig, ax = plt.subplots(figsize=(24, 10))
sns.lineplot(x='SKU', y='Lead times', data=data, ax=ax)
plt.xticks(rotation=90)
ax.set_title('Lead times by SKU')
ax.set_xlabel('SKU')
ax.set_ylabel('Lead Times')
plt.show()
```

Analyzed the stock levels , lead times and amount of revenue generated by each SKU

```python
# Plotting a pie chart between Production volumes and suppliers
mean_volumes_by_supplier = data.groupby('Supplier name')['Production volumes'].mean()
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(mean_volumes_by_supplier, labels=mean_volumes_by_supplier.index, autopct='%1.1f%%', startangle=90)
ax.set_title('Production Volumes by Supplier')
plt.show()
```

Supplier 4 gives the largest production volume while Supplies 1 gives the lowest prodcution volume

```python
# Plotting a graph between Manufacturing costs and suppliers
sns.set(style='whitegrid')
mean_costs_by_supplier = data.groupby('Supplier name')['Manufacturing costs'].mean()
sns.barplot(x=mean_costs_by_supplier.index, y=mean_costs_by_supplier.values)
plt.title('Manufacturing Costs by Supplier')
plt.xlabel('Supplier Name')
plt.ylabel('Manufacturing Costs')
plt.xticks(rotation=90)
plt.show()
```

Supplier 4 has the highest manufacturing costs while Supplier 2 has the lowest manufacturing costs

```python
# Plotting a graph between Defect rates and suppliers
sns.set(style='whitegrid')
mean_defect_rates_by_supplier = data.groupby('Supplier name')['Defect rates'].mean()
sns.barplot(x=mean_defect_rates_by_supplier.index, y=mean_defect_rates_by_supplier.values)
plt.title('Defect Rates by Supplier')
plt.xlabel('Supplier Name')
plt.ylabel('Defect Rates')
plt.xticks(rotation=90)
plt.show()
```

Supplier 5 has the maximum defect rates while supplier 1 has the minimun defect rates

```python
# Plotting a graph between shipping costs and shipping carriers
sns.set(style='whitegrid')
total_shipping_costs_by_carrier = data.groupby('Shipping carriers')['Shipping costs'].sum()
sns.barplot(x=total_shipping_costs_by_carrier.index, y=total_shipping_costs_by_carrier.values)
plt.title('Shipping Costs by Carrier')
plt.xlabel('Shipping Carrier')
plt.ylabel('Shipping Costs')
plt.xticks(rotation=90)
plt.show()
```

Carrier B has the maximum shipping costs while carrier A and carrier C have almost same shipping costs

```python
# Plotting a graph between Shipping times and shipping carriers
sns.set(style='whitegrid')
mean_shipping_times_by_carrier = data.groupby('Shipping carriers')['Shipping times'].mean()
sns.barplot(x=mean_shipping_times_by_carrier.index, y=mean_shipping_times_by_carrier.values)
plt.title('Shipping Times by Carrier')
plt.xlabel('Shipping Carrier')
plt.ylabel('Shipping Times')
plt.xticks(rotation=90)
plt.show()
```

Carrier A takes the maximum time for shipments while carrier B takes the minimum time for shipments

```python
# Plotting a graph between shipping carriers and revenue generated
sns.set(style='whitegrid')
total_revenue_by_carrier = data.groupby('Shipping carriers')['Revenue generated'].sum()
sns.barplot(x=total_revenue_by_carrier.index, y=total_revenue_by_carrier.values)
plt.title('Revenue by Shipping Carrier')
plt.xlabel('Shipping Carrier')
plt.ylabel('Revenue Generated')
plt.xticks(rotation=90)
plt.show()
```

Carrier B generated the maximum revenue while carrier A generates the minimum revenue

```python
# Plotting a graph between product types and Defect rates
sns.set(style='whitegrid')
mean_defect_rates_by_product_type = data.groupby('Product type')['Defect rates'].mean()
sns.barplot(x=mean_defect_rates_by_product_type.index, y=mean_defect_rates_by_product_type.values)
plt.title('Defect Rates by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Defect Rate')
plt.xticks(rotation=90)
plt.show()
```

Defect rates in haircare are more compared to the defect rates in cosmetics and skincare.
